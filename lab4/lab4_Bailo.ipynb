```python
import numpy as np
import matplotlib.pyplot as plt

# Дані (виберіть варіант)
X = np.array([-1, -1, 0, 1, 2, 3])  # Значення X для варіанту 1
Y = np.array([1, 0, 1, 1, 3, 5])  # Значення Y для варіанту 1

# Обчислення коефіцієнтів методом найменших квадратів
n = len(X)
X_mean = np.mean(X)
Y_mean = np.mean(Y)

# Обчислюємо β1 і β0
β1 = np.sum((X - X_mean) * (Y - Y_mean)) / np.sum((X - X_mean)**2)
β0 = Y_mean - β1 * X_mean

# Створюємо апроксимуючу функцію
Y_pred = β0 + β1 * X

# Виведення результатів
print(f"Коефіцієнти: β0 = {β0:.2f}, β1 = {β1:.2f}")
print(f"Рівняння прямої: y = {β0:.2f} + {β1:.2f}x")

# Побудова графіків
plt.scatter(X, Y, color="red", label="Експериментальні точки")
plt.plot(X, Y_pred, color="blue", label="Апроксимуюча пряма")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Лінійна регресія")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Дані для інтерполяції (приклад: варіант із X і Y)
x = np.array([0.1, 0.3, 0.4, 0.6, 0.7])  # X значення
y = np.array([3.2, 3, 1, 1.8, 1.9])      # Y значення

# Побудова інтерполяційного полінома Лагранжа
polynomial = lagrange(x, y)

# Обчислення значення функції в проміжних точках
x_test = [0.2, 0.5]
y_test = [polynomial(val) for val in x_test]

print(f"Значення функції в точках:")
for xi, yi in zip(x_test, y_test):
    print(f"f({xi}) = {yi:.2f}")

# Побудова графіка
x_dense = np.linspace(min(x), max(x), 100)  # Більше точок для гладкого графіка
y_dense = polynomial(x_dense)

plt.scatter(x, y, color="red", label="Задані точки")  # Початкові точки
plt.plot(x_dense, y_dense, color="blue", label="Інтерполяційний поліном")  # Поліном
plt.scatter(x_test, y_test, color="green", label="Проміжні точки")  # Проміжні точки
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Інтерполяція Лагранжа")
plt.legend()
plt.grid(True)
plt.show()
```
